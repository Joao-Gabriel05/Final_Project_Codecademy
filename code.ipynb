```python
data['Under18home'] = data['Kidhome'] + data['Teenhome']
data
```

```python
data['Under18home'].max()
```

```python
zerokids=data[data['Under18home']==0]
onekid=data[data['Under18home']==1]
twokids=data[data['Under18home']==2]
threekids=data[data['Under18home']==3]

onekid.head()
```

```python
axs=plt.subplot()
plt.bar(x=zerokids['Under18home'],height=zerokids['MntSweetProducts'].sum())
plt.bar(x=onekid['Under18home'],height=onekid['MntSweetProducts'].sum())
plt.bar(x=twokids['Under18home'],height=twokids['MntSweetProducts'].sum())
plt.bar(x=threekids['Under18home'],height=threekids['MntSweetProducts'].sum())
axs.set_xticks([0,1,2,3])
axs.set_xlabel('Number of kids home')
axs.set_ylabel('Total amount of sweet purchased')
plt.show()
```

```python
data['Under18home']=data['Under18home'].astype(object)
under18= data['Under18home'].value_counts().sort_index()
label = under18.index
plt.pie(under18, autopct='%1.1f%%')
plt.axis('equal') 
plt.legend(label)
plt.title('Distribution of the amount of sons by consumers')
plt.show()
```

```python
zerokids['Group'] = '0'
onekid['Group'] = '1'
twokids['Group'] = '2'
threekids['Group'] = '3'
combined_df = pd.concat([zerokids,onekid, twokids,threekids])
plt.figure(figsize=(12, 6))
sns.boxplot(data=combined_df, x='Group', y='MntSweetProducts')
plt.title('Box Plot of amount spent on sweets by the quantity of kids and teens of the user')
plt.xlabel('Quantity of kids and teens')
plt.ylabel('Amount spent in sweet')
plt.show()
```

```python
data.Country.value_counts()
```

```python
df_sp = data[data['Country']=='SP']
df_sa = data[data['Country']=='SA']
df_ca = data[data['Country']=='CA']
df_aus= data[data['Country']=='AUS']
df_ind = data[data['Country']=='IND']
df_ger = data[data['Country']=='GER']
df_us = data[data['Country']=='US']
df_me = data[data['Country']=='ME']
sum_sp = df_sp['MntFruits'].sum() + df_sp['MntWines'].sum() + df_sp['MntMeatProducts'].sum() + df_sp['MntFishProducts'].sum()+ df_sp['MntSweetProducts'].sum()
sum_sa = df_sa['MntFruits'].sum() + df_sa['MntWines'].sum() + df_sa['MntMeatProducts'].sum() + df_sa['MntFishProducts'].sum() + df_sa['MntSweetProducts'].sum()
sum_ca = df_ca['MntFruits'].sum() + df_ca['MntWines'].sum() + df_ca['MntMeatProducts'].sum() + df_ca['MntFishProducts'].sum()+ df_ca['MntSweetProducts'].sum()
sum_aus = df_aus['MntFruits'].sum() + df_aus['MntWines'].sum() + df_aus['MntMeatProducts'].sum() + df_aus['MntFishProducts'].sum() + df_aus['MntSweetProducts'].sum()
sum_ind = df_ind['MntFruits'].sum() + df_ind['MntWines'].sum() + df_ind['MntMeatProducts'].sum() + df_ind['MntFishProducts'].sum() + df_ind['MntSweetProducts'].sum()
sum_ger = df_ger['MntFruits'].sum() + df_ger['MntWines'].sum() + df_ger['MntMeatProducts'].sum() + df_ger['MntFishProducts'].sum() + df_ger['MntSweetProducts'].sum()
sum_us = df_us['MntFruits'].sum() + df_us['MntWines'].sum() + df_us['MntMeatProducts'].sum() + df_us['MntFishProducts'].sum() + df_us['MntSweetProducts'].sum()
sum_me = df_me['MntFruits'].sum() + df_me['MntWines'].sum() + df_me['MntMeatProducts'].sum() + df_me['MntFishProducts'].sum() + df_me['MntSweetProducts'].sum()
```

```python
axs=plt.subplot()
plt.bar(x=df_sp['Country'],height=sum_sp,color='orange')
plt.bar(x=df_sa['Country'],height=sum_sa,color='green')
plt.bar(x=df_ca['Country'],height=sum_ca,color='red')
plt.bar(x=df_aus['Country'],height=sum_aus,color='purple')
plt.bar(x=df_ind['Country'],height=sum_ind,color='brown')
plt.bar(x=df_ger['Country'],height=sum_ger,color='black')
plt.bar(x=df_us['Country'],height=sum_us,color='blue')
plt.bar(x=df_me['Country'],height=sum_me,color='pink')
plt.xlabel('Countries')
plt.ylabel('Total amount spent on products')
plt.title('Total amount spent on product per countries')
plt.show()
```

```python
plt.bar(x='Physical',height=data['NumStorePurchases'].sum())
plt.bar(x='Web',height=data['NumWebPurchases'].sum())
plt.ylabel('Number of purchases')
plt.title('Comparison of the number of purchases made in fisical or in web stores')
plt.show()
```

Conclusion:
The number of kids and teens at home influences the amount spent on sweets. The analysis shows that the fewer kids and teens there are at home, the more is spent on sweets.

Marital status has little influence on the amount spent on wine. The only differences observed are that widows have a higher median spending, while divorced individuals have a larger interquartile range (IQR).

The country that makes the most purchases by far is Spain. The second is Saudi Arabia, followed closely by Canada.

The purchases are mostly made in a physical store.
